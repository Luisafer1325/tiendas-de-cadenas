### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

#1. Análisis de facturación



###***Ingreso total por cada tienda***

```python
ingreso_tienda_1 = tienda["Precio"].sum()
ingreso_tienda_2 = tienda2["Precio"].sum()
ingreso_tienda_3 = tienda3["Precio"].sum()
ingreso_tienda_4 = tienda4["Precio"].sum()

```

```python
print("Ingreso tienda 1:", ingreso_tienda_1)
print("Ingreso tienda 2:", ingreso_tienda_2)
print("Ingreso tienda 3:", ingreso_tienda_3)
print("Ingreso tienda 4:", ingreso_tienda_4)
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]
```

```python
plt.figure(figsize=(9,6))
plt.bar(tiendas, ingresos, color="#7E57C2", edgecolor="black", linewidth=2)

plt.title("Ingresos Totales por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos ($)")
plt.grid(axis="y", linestyle="--", alpha=0.3)

plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def productos_mas_y_menos_vendidos(df, nombre_tienda):

    ventas = df["Producto"].value_counts()

    print(f"\nProductos MÁS vendidos en {nombre_tienda}:")
    print(ventas.head(5))

    print(f"\nProductos MENOS vendidos en {nombre_tienda}:")
    print(ventas.tail(5))


```

```python
productos_mas_y_menos_vendidos(tienda,  "Tienda 1")
productos_mas_y_menos_vendidos(tienda2, "Tienda 2")
productos_mas_y_menos_vendidos(tienda3, "Tienda 3")
productos_mas_y_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
print("Costo de envío PROMEDIO - Tienda 1:", promedio_envio_tienda_1)
print("Costo de envío PROMEDIO - Tienda 2:", promedio_envio_tienda_2)
print("Costo de envío PROMEDIO - Tienda 3:", promedio_envio_tienda_3)
print("Costo de envío PROMEDIO - Tienda 4:", promedio_envio_tienda_4)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    tienda["Precio"],
    tienda["Costo de envío"],
    color="#7E57C2",     # Color morado fuerte
    s=80,                # Tamaño de puntos
    edgecolor="black",   # Borde negro para más visibilidad
    alpha=0.8            # Un poco transparente
)

plt.title("Relación entre Precio y Costo de Envío (Tienda 1)", fontweight="bold")
plt.xlabel("Precio del Producto", fontweight="bold")
plt.ylabel("Costo de Envío", fontweight="bold")

plt.grid(linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```
